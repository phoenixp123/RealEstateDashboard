```python
import pandas as pd
import numpy as np
import plotly.offline as py
import plotly.graph_objs as go
import plotly.express as px
import datetime
```

```python
df_nat = pd.read_csv("https://econdata.s3-us-west-2.amazonaws.com/Reports/Core/RDC_Inventory_Core_Metrics_Country_History.csv")
df_state = pd.read_csv("https://econdata.s3-us-west-2.amazonaws.com/Reports/Core/RDC_Inventory_Core_Metrics_State_History.csv")
df_metro = pd.read_csv("https://econdata.s3-us-west-2.amazonaws.com/Reports/Core/RDC_Inventory_Core_Metrics_Metro_History.csv")
df_nat['month_date_yyyymm'] = pd.to_datetime(df_nat['month_date_yyyymm'], format='%Y%d')
df_state['month_date_yyyymm'] = pd.to_datetime(df_state['month_date_yyyymm'], format='%Y%d')
df_metro['month_date_yyyymm'] = pd.to_datetime(df_metro['month_date_yyyymm'], format='%Y%d')


```

```python
# simple line chart with Plotly Express
sorted_df_state = df_state.copy()

sorted_df_state["order"] = sorted_df_state["state"].map({"new york": 1, "california": 2}).fillna(3)

sorted_df_state.sort_values(by=["order","month_date_yyyymm"], ascending=False, inplace=True)


fig2 = px.line(sorted_df_state, 
        x="month_date_yyyymm", 
        y="median_listing_price", 
        color="state", 
        title="median prices over tiem by state")

# set color of all traces to lightgrey
#fig2.update_traces({"line":{"color":"lightgrey"}})

# color Turkish line to blue
fig2.update_traces(patch={"line":{"color":"blue", "width":5}}, 
                  selector={"legendgroup":"new york"})

# color Japanese line to red
fig2.update_traces(patch={"line":{"color":"red", "width":5}}, 
                  selector={"legendgroup":"california"})
# # remove the legend, y-axis and add a title
fig2.update_layout(title="Median Prices Over Time by State",
                showlegend=True)
# # plot the chart
fig2.show()
```
